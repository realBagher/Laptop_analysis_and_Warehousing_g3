```python
query = """
SELECT 
    l.Model_Name,
    l.Manufacturer,
    l.Category,
    SUM(Quantity) AS TotalQuantity,
    SUM(Profit)/SUM(Quantity) AS ProfitPerUnit,
    SUM(Profit)/SUM(Price)*100 AS ProfitPercentage
FROM `order`
JOIN laptop l ON l.laptop_id = `order`.laptop_id
GROUP BY l.Model_Name, l.Manufacturer, l.Category
ORDER BY TotalQuantity DESC
"""

Models = pd.read_sql(query, conn)
```

```python
Models.head(10)
```

```python
#Scaling
Models['TotalQuantity'] = (Models['TotalQuantity'] -Models['TotalQuantity'].min()) / (Models['TotalQuantity'].max() - Models['TotalQuantity'].min())
Models['ProfitPerUnit'] = (Models['ProfitPerUnit'] -Models['ProfitPerUnit'].min()) / (Models['ProfitPerUnit'].max() - Models['ProfitPerUnit'].min())
Models['ProfitPercentage'] = (Models['ProfitPercentage'] -Models['ProfitPercentage'].min()) / (Models['ProfitPercentage'].max() - Models['ProfitPercentage'].min())

Models
```

```python

Models['Measurement criteria'] = (Models['TotalQuantity']) + (Models['ProfitPerUnit']) +(Models['ProfitPercentage'])
best_Models = Models.sort_values(['Measurement criteria'], ascending=False)
best_Models
```

```python
best_Models = best_Models.drop(columns=['TotalQuantity', 'ProfitPerUnit','ProfitPercentage'])
```

```python
best_Models = best_Models.reset_index(drop=True)
best_Models
```
